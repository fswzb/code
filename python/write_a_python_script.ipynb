```python
#!/usr/bin/env python
#这个脚本是改写shell的脚本。从stdin中读入text,然后列出重复次数最多的单词。
# < xxx.txt | ./write_a_python_script.py 5 
```

```python
import re 
import sys
from collections import Counter
```

```python
num_words = int(sys.argv[1].strip())
#./write_a_python_scripyt.py n 。这里的n 就是num_words
text = sys.stdin.read().lower()
words = re.split('\W+', text)
cnt = Counter(words)
for word, count in cnt.most_common(num_words):    
    print "%7d %s" % (count, word)
```
